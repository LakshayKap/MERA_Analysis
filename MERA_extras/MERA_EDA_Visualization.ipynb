```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("Mera_CleanData.csv")
print("Dataset Description")
print(df1.info())
```

```python
df1.head()
```

```python
print("Summary Statistics")
df1.describe()
```

```python
df1['data'] = pd.to_datetime(df1['date'])
```

```python
# EDA: Visualizations
eda_figures = {}

# Distribution of target variable: extreme_weather
eda_figures["extreme_weather_dist"] = plt.figure(figsize=(6, 4))
sns.countplot(data=df1, x="extreme_weather", palette="viridis")
plt.title("Distribution of Extreme Events")
plt.xlabel("Extreme Weather (0: No, 1: Yes)")
plt.ylabel("Count")

# Correlation heatmap for numerical features
eda_figures["correlation_heatmap"] = plt.figure(figsize=(12, 10))
numerical_features = df1.select_dtypes(include=["float64", "int64"])
corr_matrix = numerical_features.corr()
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")

# Distribution of a numerical feature: (based on extreme_weather)
eda_figures["weather_dist"] = plt.figure(figsize=(6, 4))
sns.histplot(df1["extreme_weather"], kde=True, bins=30, color="blue")
plt.title("Distribution of Extreme Weather")
plt.xlabel("Extreme Weather (0: No, 1: Yes)")
plt.ylabel("Density")

# Boxplot of temperature by extreme_weather
eda_figures["extreme_weather_boxplot"] = plt.figure(figsize=(8, 6))
sns.boxplot(data=df1, x="extreme_weather", y="max_temp", palette="Set2")
plt.title("Temperature vs Extreme Weather")
plt.xlabel("Extreme Weather (0: No, 1: Yes)")
plt.ylabel("Temperature")

# Boxplot of precipitation by extreme_weather
eda_figures["extreme_weather_boxplot"] = plt.figure(figsize=(8, 6))
sns.boxplot(data=df1, x="extreme_weather", y="rain", palette="Set2")
plt.title("Precipitation vs Extreme Weather")
plt.xlabel("Extreme Weather (0: No, 1: Yes)")
plt.ylabel("Precipitation")


# Boxplot of wind speed by extreme_weather
eda_figures["extreme_weather_boxplot"] = plt.figure(figsize=(8, 6))
sns.boxplot(data=df1, x="extreme_weather", y="wind_speed", palette="Set2")
plt.title("Wind Speed vs Extreme Weather")
plt.xlabel("Extreme Weather (0: No, 1: Yes)")
plt.ylabel("Wind Speed")

# Pairplot of selected features (use extreme_weather as hue)
selected_features = ["rain", "pressure_cbl", "wind_speed", "temp_range", "extreme_weather"]
eda_figures["pairplot"] = sns.pairplot(df1[selected_features], hue="extreme_weather", palette="husl")

eda_figures
```
